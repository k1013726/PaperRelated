```python
# 載入需要的套件
from bs4 import BeautifulSoup
import requests
import pandas as pd
# from selenium import webdriver
# from selenium.webdriver.common.by import By
```

```python
def getherf(html):
    # 使用BeautifulSoup函式解析傳入的html
    headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/74.0.3729.169 Safari/537.36'}
    response = requests.get(html,headers=headers)
    soup = BeautifulSoup(response.text,'html.parser')
    allnode_of_a = soup.find_all("div",class_='textDiv')
    
    result = [_.get("href") for _ in allnode_of_a]
    return result
```

```python
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/74.0.3729.169 Safari/537.36'}
response = requests.get('https://www.investing.com/news/cryptocurrency-news',headers=headers)
soup = BeautifulSoup(response.text,'html5lib')
allnode_of_a = soup.find_all("a")
print(allnode_of_a)
```
